### **25. loc and iloc**

````
1. df.loc[Rows, Columns]
    - Accessing Elements through Label
    - In Slicing Using loc both start and end are included.
    - Example: df.loc[0:4, ["Name"]]
    - From 0 to 4 th Row All are Included

2. df.iloc[Rows, Columns]
    - Accessing Elements through position or index
    - In Slicing Using iloc start and end-1 are included.
    - Example: df.iloc[0:4, ["Name"]]
    - From 0 to 3 th Rows are Included
