## **Problem Statement**
- You are given two DataFrames: orders and customers.
- A customer’s order is considered shipable if the customer’s address is known (i.e., the address field is not empty and not null).

### **Task**
- Find the percentage of shipable orders.

### **DataFrame Details**

#### **orders**


| Column           | Type       |
| ---------------- | ---------- |
| id               | int64      |
| cust_id          | int64      |
| order_date       | datetime64 |
| order_details    | object     |
| total_order_cost | int64      |

#### **customers**

| Column       | Type   |
| ------------ | ------ |
| id           | int64  |
| first_name   | object |
| last_name    | object |
| city         | object |
| address      | object |
| phone_number | object |

### **Approach**

1. Merge the orders with customers using cust_id
2. Identify shipable orders where address is not null and address is not an empty string
3. Calculate percentage:
````python
percentage = (shipable orders / total orders) × 100
​

import pandas as pd

# Merge orders with customers
merged_df = orders.merge(customers, left_on='cust_id', right_on='id', how='left')

# Filter shipable orders (address not null & not empty)
shipable_df = merged_df[merged_df['address'].notna() & (merged_df['address'] != "")]

# Percentage of shipable orders
percent_shipable = (len(shipable_df) / len(merged_df)) * 100

percent_shipable
