### **8. Data Cleaning (Rank Column)**

```python
df["Rank"].unique()
```

```python
# Here Basically The Rank Column is there in Object Format We have to convert this into string.
# Because At the end If we are having data and we have to train that Machine Learning Model with Numbers
# So here i will be converting it into Number

# Step 1:
# Remove String Part
# Replace Comma 
# Convert Object to int
```

```python
df["Rank"] = df["Rank"].str[:-2]
```

```python
df["Rank"] = df["Rank"].str.replace(",","")
```

```python
df["Rank"] = df["Rank"].astype(int)
df.dtypes
```

-------------

### **9. Data Cleaning (Video Uploads and Subscribers Columns)**

```python
import pandas as pd

# Convert to integers safely
df["Subscribers"] = pd.to_numeric(df["Subscribers"], errors='coerce').astype("Int64")
df["Video Uploads"] = pd.to_numeric(df["Video Uploads"], errors='coerce').astype("Int64")

# Optional: Fill NaN with 0 if needed
# df["Subscribers"] = df["Subscribers"].fillna(0).astype(int)
# df["Video Uploads"] = df["Video Uploads"].fillna(0).astype(int)
df.dtypes
```

-------------

### **10. Data Cleaning (Grade Column)**

```python
df["Grade"].unique()
```

```python
df["Grade"] = df["Grade"].str.replace('\xa0', ' ', regex=False)
df["Grade"] = df["Grade"].str.strip()
```

```python
df["Grade"] = df["Grade"].map({"A++":5, "A+":4, "A":4, "A-":3, "B+":2})
```

```python
df.head()
```

--------------

### **11. Find Average Views for Each Channel**

```python
df["Video Uploads"] = pd.to_numeric(df["Video Uploads"], errors = 'coerce')
```

```python
df["Average_Views"] = df["Video views"]/ df["Video Uploads"]
df
```

### **14. Which Grade Has Maximum Number of Video Uploads**

```python
df1 = df.groupby("Grade")["Video Uploads"].sum().sort_values(ascending=False)
df1
```

```python
# By Using Plot

import matplotlib.pyplot as plt

plt.bar(df1.index.astype(str), df1.values)
```

### **15. Which Grade Has the Highest Average Views**

```python
df.columns
```

```python
import matplotlib.pyplot as plt

# Calculate average views per grade
df_avg = df.groupby("Grade")["Average_Views"].mean().sort_values(ascending=False)
df_avg
```
